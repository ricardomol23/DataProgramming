### Evaluo los valores faltantes
Para este proceso creo un nuevo DF llamado missing_values_isnull y la funcion .isnull para poder saber los valores nulos

```python
missing_data = red.isnull()
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

### Conclusión
En este caso podemos, observar que no tenemos valores faltantes
