```python
import pandas as pd

# Suponiendo que 'red' es tu DataFrame cargado con los datos

# Análisis descriptivo de variables numéricas
numeric_variables = ['year', 'selling_price', 'km_driven', 'mileage', 'engine', 'max_power', 'seats']
numeric_stats = red[numeric_variables].describe()
print("Estadísticas descriptivas de variables numéricas:")
print(numeric_stats)

# Análisis descriptivo de variables categóricas
categorical_variables = ['Region', 'State or Province', 'fuel', 'seller_type', 'transmission', 'owner', 'sold']
for column in categorical_variables:
    print("\nConteo de valores únicos para la variable:", column)
    print(red[column].value_counts())
```

```python

# Define thresholds for low, medium, and high based on your data distribution
low_threshold = np.percentile(red["max_power"], 33)  # 33rd percentile (low)
medium_threshold = np.percentile(red["max_power"], 66)  # 66th percentile (medium)
high_threshold = red["max_power"].max()  # Maximum value (high)

# Create bins based on the thresholds
bins = np.linspace(0, high_threshold, 4)  # 4 bins for low, medium, high + overflow
```

```python
# Define group names for the bins
group_names = ["Low HP POWER", "Medium HP POWER", "High HP POWER"]  # Adjust names as needed

# Create the new column with horsepower categories (correcting the typo)
red['HP - Group'] = pd.cut(red['max_power'], bins=bins, labels=group_names, include_lowest=True)
```

```python
# Count the occurrences of each category in the "HP - Groups" column
red['HP - Group'].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, red["HP - Group"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("max_power")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Suponiendo que 'red' es tu DataFrame cargado con los datos

# Relación entre Kilometraje y Precio de Venta
plt.figure(figsize=(10, 6))
sns.scatterplot(x='km_driven', y='selling_price', data=red)
plt.title('Relación entre Kilometraje y Precio de Venta')
plt.xlabel('Kilometraje')
plt.ylabel('Precio de Venta')
plt.show()

# Probabilidad de Venta en función del Kilometraje
red['mileage_group'] = pd.cut(red['mileage'], bins=5)  # Divide el kilometraje en 5 grupos
mileage_sales = red.groupby('mileage_group')['sold'].mean()
plt.figure(figsize=(10, 6))
mileage_sales.plot(kind='bar')
plt.title('Probabilidad de Venta en función del Kilometraje')
plt.xlabel('Grupo de Kilometraje')
plt.ylabel('Probabilidad de Venta')
plt.xticks(rotation=45)
plt.show()

# Comparación de Kilometraje entre Vehículos Vendidos y No Vendidos
plt.figure(figsize=(10, 6))
sns.boxplot(x='sold', y='km_driven', data=red)
plt.title('Comparación de Kilometraje entre Vehículos Vendidos y No Vendidos')
plt.xlabel('Vendido')
plt.ylabel('Kilometraje')
plt.xticks([0, 1], ['No Vendido', 'Vendido'])
plt.show()
```
